# Strings

<div style="text-align: justify"> 
Strings are used to store text data in python
    </div>

```python
text = 'python tutorial'
text
```

```python
# access the first character 'p'
text[0]
```

```python
# Strings are immutable (you cannot change their charaters)
# run the code and see an error
text[1]='d'
```

```python
# access substrings 'tutorial'
text[7:15]
```

## Multiple lines

```python
# write multiple lines with '''
lines = '''python basices
for
Data Science'''
print(lines)
```

## Concatenation

```python
# use + sign to concatenate two strings
s1 = "Data"
s2 = "Science"
s1 + " " + s2
```
