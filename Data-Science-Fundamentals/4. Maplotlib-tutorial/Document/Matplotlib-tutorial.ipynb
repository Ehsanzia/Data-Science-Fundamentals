# Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,53,47,49,45,43]
```

```python
plt.plot(x,y,color='green',linewidth=3,linestyle='dashed')
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')
```

Tip: **If you get this error by using plt.xlabel <font color='red'>TypeError: 'str' object is not callable</font> restart your kernel**

# Format string in plot option

```python
plt.plot(x,y,'b--.')
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')
```

```python
plt.plot(x,y,'--*r')
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')
```

# Scale factor
 0 $<$ alpha $<$ 1

```python
plt.plot(x,y,'b--.',alpha=0.6)
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('weather')
```

[link for additional options on plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

## Legend

```python
days = [1,2,3,4,5,6,7]
max_t = [51,50,53,47,49,47,45]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend(loc='best',shadow=True,fontsize='large')
# loc='best' i.e. matplotlib module decide where is the best place for legend.
plt.grid()
```

[Chech this link for more information](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

## Bar chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company =['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bin)')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Revenue')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bin)')
plt.title('US Tech Stocks')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.ylabel('revenue(bin)')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.title('US Tech Stocks')
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()
```

## Histogram

1. How many patients are normal?

2. How many of them are pre-diabetic?

3. How many are diabetic?

| blood-sugar        | patient status|
|--------------------|---------------|
| 80-100             | Normal        |
| 100-125            | Pre-diabetic  |
| above 125          | Diabetic      | 


```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.9,color='g')
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.9,color=['b','r'],
         label=['men','women'])
plt.legend()
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.9,color=['b','r'],
         label=['men','women'],orientation='horizontal')
plt.legend()
```

# Pie chart

In order to track down the home expences I want to know how much percentage individual item is taking out of total expense having a pie chart is useful.


| expense label     | expense value      |
|--------------------|---------------|
| Home Rent   |  1490    |
| Food             | 1400        |
| Phone            | 600  |
| Car          | 300      |
| Other Utilities | 250    |
| Total  | 2960 |


```python
exp_vals = [1490,1400,600,300,250]
exp_labels = ['Home Rent', 'Food','Phone','Car','Other Utilities']
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=[0,0.1,0.2,0,0.3])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',startangle=45)
plt.show()
```

# Save Chart to a file

```python

plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',startangle=45)
plt.savefig('D:/Data_Science/My Github/Maplotlib-tutorial/Pictures/piechart.png',
           bbox_inches='tight',pad_inches=2,transparent=True)
```
